```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data2 = pd.read_csv("us_cities_small.csv")
data1= pd.read_csv("user_data.csv")
data = pd.merge(data2,data1)
data.loc[data.population_proper <100000, "location"]="rural"
data.loc[data.population_proper>=100000, "location"]="urban"
```

```python
sns.regplot(x = "population_proper",y="age",data=data)
plt.show()
```

```python
plt.close()

sns.set_style("darkgrid")
sns.set_palette("bright")
sns.despine()

sns.regplot(x = "population_proper", y = "age", data = data)
ax = plt.subplot(1,1,1)
ax.set_xticks([100000, 1000000, 2000000, 4000000, 8000000])
ax.set_xticklabels(["100k", "1m", "2m", "4m", "8m"])

plt.show()

```

```python
plt.close()
sns.regplot(x = "population_proper", y = "age", data= data)
ax = plt.subplot(1,1,1)
ax.set_xticks([100000, 1000000, 2000000, 4000000, 8000000])
ax.set_xticklabels(["100k", "1m", "2m", "4m", "8m"])
ax.set_xlabel("City Population")
ax.set_ylabel("User Age")
```
